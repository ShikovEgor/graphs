```python
import pandas as pd
import numpy as np
```

```python
edges = pd.read_csv('cora.cites', sep='\t', header=None,names=["Target", "Source"])
```

```python
edges = edges.append(edges.rename(columns={'Target':'Source', 'Source':'Target'})).drop_duplicates()
```

```python
feature_names = ["w_{}".format(ii) for ii in range(1433)]
column_names =  feature_names + ["subject"]
nodes = pd.read_csv("cora.content", sep='\t', header=None, names=column_names)
```

```python
nodes.subject
```

```python
from sklearn import preprocessing
```

```python
le = preprocessing.LabelEncoder()
```

```python
le.fit(nodes.subject)
```

```python
nodes.subject = le.transform(nodes.subject)
```

```python
nodes
```

```python
features = dict(zip(nodes.index,[l.tolist() for l in nodes.values]))
```
